```python
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import SpectralClustering
import matplotlib.pyplot as plt

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data

# Perform Spectral Clustering
n_clusters = 2  # Specify the number of clusters
spectral_clustering = SpectralClustering(n_clusters=n_clusters, affinity='nearest_neighbors', random_state=42)
cluster_labels = spectral_clustering.fit_predict(X)

# Visualize the clusters (assuming you can only plot 2 features)
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='viridis', alpha=0.5)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Spectral Clustering')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Compute Silhouette Score
silhouette = silhouette_score(X, cluster_labels)
print(f'Silhouette Score: {silhouette}')

# Compute Davies-Bouldin Index
db_score = davies_bouldin_score(X, cluster_labels)
print(f'Davies-Bouldin Index: {db_score}')

# Compute Calinski-Harabasz Index
calinski_score = calinski_harabasz_score(X, cluster_labels)
print(f'Calinski-Harabasz Index: {calinski_score}')
```
