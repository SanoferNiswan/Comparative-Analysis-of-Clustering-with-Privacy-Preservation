```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
```

```python
# Load the dataset
wine_data = pd.read_csv('winequality-red.csv')
```

```python
# Standardize the features
scaler = StandardScaler()
wine_data_scaled = scaler.fit_transform(wine_data)
```

```python
# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
cluster_labels = dbscan.fit_predict(wine_data_scaled)
```

```python
# Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=wine_data_scaled[:, 0], y=wine_data_scaled[:, 1], hue=cluster_labels, palette='viridis')
plt.title('DBSCAN Clustering of Wine Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# Calculate performance metrics
silhouette = silhouette_score(wine_data_scaled, cluster_labels)
davies_bouldin = davies_bouldin_score(wine_data_scaled, cluster_labels)
calinski_harabasz = calinski_harabasz_score(wine_data_scaled, cluster_labels)
```

```python
print("Silhouette Score:", silhouette)
print("Davies-Bouldin Index:", davies_bouldin)
print("Calinski-Harabasz Index:", calinski_harabasz)
```
