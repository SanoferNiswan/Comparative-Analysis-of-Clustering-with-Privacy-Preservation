```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Load the Wine Quality dataset (assuming it's a CSV file)
wine_data = pd.read_csv("winequality-red.csv")
```

```python
# Separate features (attributes) from the target variable (quality)
X = wine_data.drop(columns='quality')  # Features
y = wine_data['quality']                # Target
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply PCA
pca = PCA(n_components=2)  # We'll reduce the data to 2 dimensions for visualization
X_pca = pca.fit_transform(X_scaled)
```

```python
# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance_ratio)
```

```python
# Visualize Results
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', alpha=0.8)
plt.title('PCA of Wine Quality Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Quality')
plt.grid(True)
plt.show()
```
