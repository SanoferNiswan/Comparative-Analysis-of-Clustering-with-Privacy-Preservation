```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Encode categorical variables
label_encoder = LabelEncoder()
categorical_columns = bd.select_dtypes(include=['object']).columns
for col in categorical_columns:
    bd[col] = label_encoder.fit_transform(bd[col])
```

```python
# Standardize numerical features (optional)
scaler = StandardScaler()
numerical_columns = bd.select_dtypes(include=['int64', 'float64']).columns
bd[numerical_columns] = scaler.fit_transform(bd[numerical_columns])
```

Exploratory Data Analysis (EDA):

```python
import matplotlib.pyplot as plt
import seaborn as sns

# View basic statistics
print(bd.describe())
```

```python
# Visualize distribution of numerical features
bd[numerical_columns].hist(bins=20, figsize=(12, 10))
plt.show()
```

```python
# Visualize counts of categorical features
plt.figure(figsize=(12, 8))
sns.countplot(data=bd, x='job')
plt.xticks(rotation=45)
plt.show()
```

```python
# Explore correlations between features
plt.figure(figsize=(12, 10))
sns.heatmap(bd.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Explore relationships between features using pair plot
sns.pairplot(bd, vars=numerical_columns, hue='y')
plt.show()
```

```python
# Visualize class distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=bd, x='y')
plt.show()
```
